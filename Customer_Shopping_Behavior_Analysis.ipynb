```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_group

labels = ["Young Adult","Adult","Middle-aged","Senior"]
df["age_group"] = pd.qcut(df["age"],q=4,labels=labels)  

# pd.qcut() is a function Divides the data (age) into 4 quantile segmenmts and assign labels (list of labels) to them.   
```

```python
df[["age","age_group"]].head(10)
```

```python
# Create column purchase_frequency_Days

# Creating a dictionary containing frequency_mapping values to convert frequency_of_purchases column into numeric to make it easy to analyse
frequency_mapping = {
    "Fortnightly" : 14,
    "Weekly" : 7,
    "Monthly" : 30,
    "Quarterly" : 90,
    "Bi-Weekly" : 14,
    "Annually" : 365,
    "Every 3 Months" : 90
}

df["purchase_frequency_Days"] = df["frequency_of_purchases"].map(frequency_mapping)  # map function is used to replace the text with numeric in a new column for containing df["frequency_of_purchases"]
```

```python
df[["frequency_of_purchases","purchase_frequency_Days"]].head(10)
```

```python
df[["discount_applied","promo_code_used"]].head(10)  # These both columns looks same so need to check if they are actually similar.
```

```python
(df["discount_applied"] == df["promo_code_used"]).all()  # "==" operator to check if they are similar or not and all() is used to get all data at once.

# True means both of them carry exactly same information
```

```python
df = df.drop("promo_code_used", axis=1)  # Dropping the column "promo_code_used" with axis=1 means dropping the whole column.
```
